### Step 7 - Convert Date Column

```python
# Convert Date Column
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
```

### Step 8 - Create Age Column


```python
### Create Age Column
df["Age"] = 2014 - df["Year_Birth"]
```

### Step 9 - Create Age Group

```python
df["Age Group"] = pd.cut(
    df["Age"],
    bins = [0, 25, 40, 60, 100], 
    labels = ["Young", "Adult", "Middle Age", "Senior"])
```

### Step 10 - Create Income Group

```python
df["Income Group"] = pd.cut(
    df["Income"],
    bins = [0, 30000, 70000, 120000, 700000], 
    labels = [
        "Low Income", 
        "Middle Income",
        "Upper Middle Income",
        "High Income"
    ]
)
```

### Step 11 - Create Total Spend

```python
# Creating Total Spend
df["Total Spend"] = (
    df['MntWines'] + 
    df['MntMeatProducts'] + 
    df['MntFruits'] + 
    df['MntFishProducts'] + 
    df['MntSweetProducts'] + 
    df['MntGoldProds']
)
```

###  Step 12 - Create Family Size 

```python
# Creating Family Size
df["Family Size"] = (
    np.where(
        df["Marital_Status"].isin(["Married", "Together"]),
        2,
        1
    )
    + df["Kidhome"]
    + df["Teenhome"]
)
```

### Create Category Breadth

```python
# How many product categories a customer purchased from.
df['Category_Breadth'] = (
    df[['MntWines',
        'MntFruits',
        'MntMeatProducts',
        'MntFishProducts',
        'MntSweetProducts',
        'MntGoldProds']] > 0
).sum(axis=1)
```

### Step 13 - Univariate Analysis

```python
# Income Distribution
plt.figure(figsize=(8,5))
plt.hist(df["Income"])
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Count")
plt.show()
```

### Step 14 - Age Group Analysis

```python
df["Age Group"].value_counts().plot(kind = "bar")
plt.title("Customers by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

### Step 15 - Total Spend Analysis

```python
plt.figure(figsize=(8,5))
plt.hist(df["Total Spend"])
plt.title("Total Spend Distribution")
plt.xlabel("Total Spend")
plt.ylabel("Count")
plt.show()
```

### Step 16 - Bivariate Analysis

### Income vs Spend Scatter Plot

```python
# Income vs Spend Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df["Income"], df["Total Spend"])

plt.xlabel("Income")
plt.ylabel("Total Spend")
plt.title("Income vs Total Spend")

plt.show()
```

### STEP 17 — Purchase Channel Analysis

```python
# Purchase Channel Analysis
channels = [
    'NumWebPurchases',
    'NumStorePurchases',
    'NumCatalogPurchases',
    'NumDealsPurchases'
]

channel_totals = df[channels].sum()

channel_totals.plot(kind='bar')
plt.title("Purchase Channels")
plt.xlabel("Channel")
plt.ylabel("Purchases")
plt.show()
```

### STEP 18 — Response Analysis

```python
# Create Response Rate
df['Response'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Campaign Response")
plt.ylabel("")
plt.show()
```

### Step 19 - Boxplot — Income Outliers

```python
plt.figure(figsize=(8,5))

plt.boxplot(
    df['Income'],
    patch_artist=True,
    boxprops=dict(facecolor='lightblue'),
    medianprops=dict(color='red')
)

plt.title("Income Boxplot")
plt.ylabel("Income")

plt.show()
```

### total purchase

```python
# total purchase 
df['Total_Purchases'] = (
    df['NumWebPurchases'] +
    df['NumCatalogPurchases'] +
    df['NumStorePurchases'] +
    df['NumDealsPurchases']
)
```

### Create Web Purchase Ratio

```python
df['Web_Purchase_Ratio'] = (
    df['NumWebPurchases'] /
    df['Total_Purchases']
)
```
